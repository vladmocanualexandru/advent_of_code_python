```python
def number_is_invalid(number):
    l10 = math.floor(math.log10(number)) # biggest power of ten to fit in number
    if l10 %2 == 0: # number cannot be halfed 
        return False
    
    halfL10 = math.pow(10,(l10+1)/2) # power of ten to split number in half
    numberLeft = math.floor(number/halfL10) # get left half of number
    numberRight = number % halfL10 # get right half of number

    return numberLeft==numberRight 
```

```python
part1Data = originalData
log(part1Data[:10])
```

```python
part1Result = 0

for numberPair in part1Data:
    number = numberPair[0]

    while number<=numberPair[1]:
        if number_is_invalid(number):
            part1Result+=number
            
        number+=1

log(yellow(part1Result))
```

```python
def number_is_invalid(number):
    number = str(number)

    for sqLen in range(1, math.floor(len(number)/2)+1):
        if not len(number) % sqLen == 0:
            continue
        sq = number[0:sqLen]

        invalid = False
        for checkPos in range(sqLen, len(number), sqLen):
            if sq != number[checkPos:checkPos+sqLen]:
                invalid = True
                break

        if not invalid:
            return True

    return False
```
