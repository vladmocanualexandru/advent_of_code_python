```python
data = getInputData(DATA_FILENAME)
```

```python
import copy

def recCountPaths(mapping, node, target, visitedNodes):
    if node == target:
        # print(visitedNodes)
        return 1

    paths = 0
    for nextNode in mapping[node]["connections"]:
        if not nextNode in visitedNodes: 
            if mapping[nextNode]["pathsToTarget"] == None:
                # print(f"node {nextNode} is not processed - processing...")
                localPaths = recCountPaths(mapping, nextNode, target, visitedNodes+[nextNode])
                mapping[nextNode]["pathsToTarget"]= localPaths
                # print(f"node {nextNode} has {localPaths} paths")
                paths+=localPaths
            else:
                # print(f"node {nextNode} processed - using value {mapping[nextNode]["pathsToTarget"]}")
                paths += mapping[nextNode]["pathsToTarget"]
        
    mapping[node]["pathsToTarget"] = paths
    return paths


svr2dac = recCountPaths(copy.deepcopy(data), "svr", "dac", ["svr"])
dac2fft = recCountPaths(copy.deepcopy(data), "dac", "fft", ["dac"])
fft2out = recCountPaths(copy.deepcopy(data), "fft", "out", ["fft"])

svr2fft = recCountPaths(copy.deepcopy(data), "svr", "fft", ["svr"])
fft2dac = recCountPaths(copy.deepcopy(data), "fft", "dac", ["fft"])
dac2out = recCountPaths(copy.deepcopy(data), "dac", "out", ["dac"])

result = svr2dac*dac2fft*fft2out + svr2fft*fft2dac*dac2out
print(result)
```
