```python
def getInputData(filename):
    

    lightPatterns = []
    buttonConfigs = []
    joltages = []

    for entry in getTuples_text(filename,' '):
        lightPatterns.append([1 if l == '#' else 0  for l in entry[0][1:-1]])

        buttonConfig = []
        for buttonConfigRaw in entry[1:-1]:
            buttonConfig.append([int(button) for button in buttonConfigRaw[1:-1].split(',')])
        buttonConfigs.append(buttonConfig)

        joltages.append([int(jlt) for jlt in entry[-1][1:-1].split(',')])

    

    return (lightPatterns, buttonConfigs, joltages)
```
