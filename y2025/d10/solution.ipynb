```python
def getInputData(filename):
    

    lightPatterns = []
    buttonConfigs = []
    joltages = []

    for entry in getTuples_text(filename,' '):
        lightPatterns.append([1 if l == '#' else 0  for l in entry[0][1:-1]])

        buttonConfig = []
        for buttonConfigRaw in entry[1:-1]:
            buttonConfig.append([int(button) for button in buttonConfigRaw[1:-1].split(',')])
        buttonConfigs.append(buttonConfig)

        joltages.append([int(jlt) for jlt in entry[-1][1:-1].split(',')])

    

    return (lightPatterns, buttonConfigs, joltages)
```

```python
def testButtons(lp, bcs, presses):
    lp = []+lp
    for press in presses:
        for b in bcs[press]:
            lp[b] = 1-lp[b]

    return sum(lp) == 0

def recChoosePresses(lp, presses, moves, bcs):
    if len(presses) == moves:
        if testButtons(lp, bcs, presses):
            return presses
        else:
            return None
    else:
        for i in range(len(bcs)):
            result = recChoosePresses(lp, presses+[i], moves, bcs)
            if result != None:
                return result
    return None


    
```

```python
(lightPatterns, buttonConfigs, joltages) = getInputData(DATA_FILENAME)

result = 0
for mi in range(len(lightPatterns)):
    lp = lightPatterns[mi]
    bc = buttonConfigs[mi]

    moves = 1
    while(True):
        presses = recChoosePresses(lp, [], moves, bc)
        if presses != None:
            result+=len(presses)
            break
        moves+=1

print(result)

```
