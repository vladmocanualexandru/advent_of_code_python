PART 2

```python
map = matrixUtils.clone(originalData)

startMs = round(time.time()*1000)

paperCoords = matrixUtils.find(map, returnCondition=(lambda e:e=='@'))
part2Result = 0

goAgain = True
while goAgain:
    goAgain = False

    newPaperCoords = []
    for (y,x) in paperCoords:
        if sum([1 if e == '@' else 0 for e in matrixUtils.getNeighbors8(map, y, x)])<4:
            goAgain = True
            map[y][x] = '.'
            part2Result+=1
        else:
            newPaperCoords.append((y,x))

    paperCoords = newPaperCoords

stopMs = round(time.time()*1000)
print(f"{yellow(part2Result)} in {stopMs-startMs}ms")
```
