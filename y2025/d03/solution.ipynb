```python
def calculate_CMDN(arr, length): # Consecutive Maximum Digit Number
    result = 0
    for l in range(length-1, 0, -1):
        result*=10

        leftMax = max(arr[:0-l])
        leftMaxIndex = arr.index(leftMax)

        arr = arr[leftMaxIndex+1:]

        result+=leftMax

    result = result*10+max(arr)
    return result
```
