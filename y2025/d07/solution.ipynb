PART 1

```python
def part1(data):
    beamStatus = [False for i in range(len(data[0]))]
    startPos = matrixUtils.find(data[:1], returnCondition=lambda e:e=='S')

    beamStatus[startPos[0][1]] = True

    result = 0
    for line in data[2:]:
        for beamSplit in matrixUtils.find(line, returnCondition=lambda e:e=='^'):
            splitPos = beamSplit[0]
            if beamStatus[splitPos]:
                result+=1
                beamStatus[splitPos-1] = True
                beamStatus[splitPos] = False
                beamStatus[splitPos+1] = True
                
    return result

data = getInputData(DATA_FILENAME)
print(yellow(part1(data)))
```

PART 2

```python
def part2(data):
    pathCount = [0 for i in range(len(data[0]))]
    startPos = matrixUtils.find(data[:1], returnCondition=lambda e:e=='S')

    pathCount[startPos[0][1]] = 1

    for line in data[2:]:
        for beamSplit in matrixUtils.find(line, returnCondition=lambda e:e=='^'):
            splitPos = beamSplit[0]
            if pathCount[splitPos]!=0:
                pathCount[splitPos-1] += pathCount[splitPos]
                pathCount[splitPos+1] += pathCount[splitPos]
                pathCount[splitPos] = 0

    return sum(pathCount)

data = getInputData(DATA_FILENAME)
print(yellow(part2(data)))
```
