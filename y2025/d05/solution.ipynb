```python
def getInputData(filename):
    allLines = getStrings(filename)

    ranges = []
    ids = None
    for line in allLines:
        if line.strip() == '':
            ids = []
            continue

        if ids == None:
            ranges.append([int(t) for t in line.split('-')])
        else:
            ids.append(int(line))
    
    return (ranges, ids)
```
