```python
def getInputData(inputFile):
    data = getRawText(inputFile)
    return data
```

```python
data = getInputData("test.data")
```

PART 1

```python
def clean_data(df):
    df = df[df.apply(lambda e: e[0][0]=='(', axis=1)] # remove entries that do not start with (
    df.loc[:, 0] = df.apply(lambda e: e[0].split(')')[0][1:], axis=1) # keep only part before ), remove (
    df.loc[:, 0] = df.apply(lambda e: e[0].split(','), axis=1) # split by comma (only the two numbers should remain)
    df = df[df.apply(lambda e: len(e[0])==2 and e[0][0].isnumeric() and e[0][1].isnumeric(), axis=1)] # keep only entries with 2 numbers
    return df
```

```python
part1Data = data.split("mul")
df = clean_data(pd.DataFrame(part1Data))

part1Result = np.sum(df.apply(lambda e: int(e[0][0])*int(e[0][1]), axis=1))
print(part1Result)
```

PART 2

```python
dontBlocks = data.split("don't()") # split into don't blocks
part2Data = dontBlocks[0] # first one is always relevant

for block in dontBlocks[1:]:
    doBlocks = block.split("do()") # split each don't block into do blocks 
    if len(doBlocks)>1: # first one is not relevant
        part2Data+="".join(doBlocks[1:]) # join and save other blocks

part2Data = part2Data.split("mul") # same as part 1
df = clean_data(pd.DataFrame(part2Data))

part2Result = np.sum(df.apply(lambda e: int(e[0][0])*int(e[0][1]), axis=1))
print(part2Result)
```
