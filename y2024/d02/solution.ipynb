```python
data = getInputData("test.data")
```

PART 1

```python
def report_is_safe(t):
    asc = t[0]<t[1]
    for i in range(len(t)-1):
        diff = t[i]-t[i+1]
        if asc != (diff<0) or not abs(diff) in [1,2,3]:
            return False
    return True
```

```python
unsafeData = []

count = 0
for t in data:
    if not report_is_safe(t):
        unsafeData.append(t) 

part1Result = len(data) - len(unsafeData)
part1Result
```

PART 2

```python
count = 0
for t in unsafeData:
    for i in range(len(t)):
        if report_is_safe(t[0:i]+t[i+1:len(t)]):
            count+=1
            break

part2Result = count+part1Result
part2Result
```
