```python
def getInputData(filename):
    raw = getStrings(filename)

    rules = []
    updates = []
    
    addingRules = True
    for entry in raw:
        if addingRules and entry.strip() == "":
            addingRules = False
            continue

        if addingRules:
            rules.append(entry)
        else:
            updates.append(entry.split(","))
    
    return (rules,updates)
```

PART 1

```python
(rules,updates) = originalData

log(rules[:10])
log(updates[:10])
```

```python
part1Result = 0

invalidUpdates = []

for update in updates:
    updateIsValid = True
    for i in range(len(update)-1):
        for j in range(i+1, len(update)):
            if not f"{update[i]}|{update[j]}" in rules:
                updateIsValid = False
                break
        if not updateIsValid:
            break
    
    if updateIsValid:
        part1Result += int(update[math.floor(len(update)/2)])
    else: 
        invalidUpdates.append(update)

log(yellow(part1Result))
```

PART 2

```python
from functools import cmp_to_key

def cmp(a,b):
    if a==b: return 0

    if f"{a}|{b}" in rules:
        return -1
    else:
        return 1
```

```python
part2Result = 0

for update in invalidUpdates:
    update.sort(key=cmp_to_key(cmp))

    part2Result += int(update[math.floor(len(update)/2)])

log(yellow(part2Result))
```
