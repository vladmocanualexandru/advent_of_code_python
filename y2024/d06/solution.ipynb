```python
map = matrixUtils.clone(part1Data)

directions = [(-1,0), (0,1), (1,0), (0,-1)]
direction = 0

(guardY, guardX) = matrixUtils.find(map,returnCondition=(lambda e:e=='^'))[0]
map[guardY][guardX] = 'X'

guardLeft = False
while not guardLeft:
    (nextGuardY, nextGuardX) = (guardY+directions[direction][0], guardX+directions[direction][1])

    if (nextGuardY < 0 or nextGuardX<0 or nextGuardY==len(map) or nextGuardX==len(map[0])): # guard left
        guardLeft = True
        break

    if map[nextGuardY][nextGuardX] == '#':
        direction=(direction+1)%4
        continue

    (guardY, guardX) = (nextGuardY, nextGuardX)
    map[guardY][guardX] = 'X'

part1Result = matrixUtils.addAll(map, lambda e:e=="X", lambda e:1)

print(yellow(part1Result))
```
