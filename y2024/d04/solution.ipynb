PART 1

```python
def search_horizontally(matrix):
    count = 0

    for row in matrix:
        colPos = 0

        while colPos < len(row) - 3:
            letters = row[colPos:colPos+4]

            if "".join(letters) == "XMAS" or "".join(letters[::-1]) == "XMAS":
                count+=1
                colPos+=3
            else:
                colPos+=1

    return count


def search_main_diagonal(matrix):
    count = 0

    for diagIndex in [0]:
        lineIndex = diagIndex
        colIndex = 0

        step = 0
        while step < len(matrix)-3-diagIndex:
            letters = [matrix[colIndex][lineIndex], matrix[colIndex+1][lineIndex+1], matrix[colIndex+2][lineIndex+2], matrix[colIndex+3][lineIndex+3]] 

            if "".join(letters) == "XMAS" or "".join(letters[::-1]) == "XMAS":
                count+=1

                lineIndex+=3
                colIndex+=3
                step+=3
            else:
                lineIndex+=1
                colIndex+=1
                step+=1

    return count

def search_under_main_diagonal(matrix):
    count = 0

    for diagIndex in range(1,len(matrix)-3):

        # di=0, 0,0 -> 6,6
        # di=1, 1,0 -> 6,5
        # di=2, 2,0 -> 6,4
        # di=k, k,0 -> 6,6-k

        # di=6, 6,0 -> 6,0

        lineIndex = diagIndex
        colIndex = 0

        step = 0
        while step < len(matrix)-3-diagIndex:
            letters = [matrix[colIndex][lineIndex], matrix[colIndex+1][lineIndex+1], matrix[colIndex+2][lineIndex+2], matrix[colIndex+3][lineIndex+3]] 

            if "".join(letters) == "XMAS" or "".join(letters[::-1]) == "XMAS":
                count+=1

                lineIndex+=3
                colIndex+=3
                step+=3
            else:
                lineIndex+=1
                colIndex+=1
                step+=1

    return count
```

```python
part1Data = originalData
```

```python
part1Result = search_horizontally(part1Data)

part1Data = matrixUtils.rotate(part1Data)
part1Result += search_horizontally(part1Data)

part1Result += search_main_diagonal(part1Data)

part1Data = matrixUtils.rotate(part1Data)
part1Result += search_main_diagonal(part1Data)

part1Result += search_under_main_diagonal(part1Data)

part1Data = matrixUtils.rotate(part1Data)
part1Result += search_under_main_diagonal(part1Data)

part1Data = matrixUtils.rotate(part1Data)
part1Result += search_under_main_diagonal(part1Data)

part1Data = matrixUtils.rotate(part1Data)
part1Result += search_under_main_diagonal(part1Data)


log(yellow(part1Result))
```
